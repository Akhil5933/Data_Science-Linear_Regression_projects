```python
#splitting of data to train the model into x(independent variable) and y(dependent variable) 
x=ds[['income']]
y=ds[['happiness']]
```

```python
#model validation process which makes dividing of the data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# importing matplotlib forplotting the graphs and naming
import matplotlib.pyplot as plt
plt.scatter(x,y,color='red')
plt.xlabel('income')
plt.ylabel('happiness')
```

```python
#importing the model for evaluation process 
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
model.predict(x_train)
```

```python
#comparing the values of training and test to get evaluated score from it, which shows performance of model for a dataset.
model.score(x_test,y_test)
```

```python
#the data points which are near to the predicted one.
plt.scatter(x,y,color='green')
plt.plot(x,model.predict(x),color='red')
```

```python
# the above model is better model(>50%) and it is not best one because it is not reaching as much as high(>80% or 90%).
# we can improve the above model by applying different regression algorithms and changing the test_size etc.
```

```python
#if the line passes more data points means it has residual is less and cost function is less.
# in this case the data points are close to the predicted line are less so accuracy of model is less.
#it is not a best fit line as the data points are less close.
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_pred)
r_squared = r2_score(y_test, y_pred)
```

```python
print(f"Mean Squared Error (MSE): {mse:f}")
print(f"R-squared: {r_squared:.2f}")
```
